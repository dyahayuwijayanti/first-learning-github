# Data Analysis Introduction

This notebook demonstrates basic data analysis using Python and pandas.


```python
import pandas as pd
```

## Create Sample Dataset


```python
data = {
    "Product": ["A", "B", "C", "D"],
    "Sales": [120, 90, 150, 80],
    "Month": ["Jan", "Jan", "Jan", "Jan"]
}

df = pd.DataFrame(data)
df
```

## Basic Analysis


```python
df.describe()
```

## Data Filtering


```python
df[df["Sales"] > 100]
```

## Conclusion

## Load Dataset from CSV


```python
df = pd.read_csv("data-analysis-basic/sample_sales.csv")
```

## Data Visualization


```python
import matplotlib.pyplot as plt

plt.bar(df["Product"], df["Sales"])
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Sales per Product")
plt.show()
```

## Insight

From the bar chart, Product C has the highest sales, while Product D has
the lowest sales.
