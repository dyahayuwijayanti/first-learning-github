# Exploratory Data Analysis (EDA)

This notebook explores a real dataset to understand patterns,
distributions, and relationships between variables.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset


```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
df.head()
```

## Dataset Overview


```python
df.info()
df.describe()
```

## Distribution of Total Bill


```python
plt.hist(df["total_bill"], bins=10)
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("Distribution of Total Bill")
plt.show()
```

## Relationship between Total Bill and Tip


```python
plt.scatter(df["total_bill"], df["tip"])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Total Bill vs Tip")
plt.show()
```

## Insights

- Higher total bills tend to result in higher tips
- Most total bill values are concentrated below 30
- The dataset is clean and suitable for basic analysis
