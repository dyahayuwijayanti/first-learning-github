# Linear Regression Introduction

This notebook demonstrates a simple linear regression example
using scikit-learn.


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Create Sample Dataset


```python
data = {
    "StudyHours": [1, 2, 3, 4, 5],
    "Score": [50, 55, 65, 70, 80]
}

df = pd.DataFrame(data)
df
```

## Prepare Data


```python
X = df[["StudyHours"]]
y = df["Score"]
```

## Train Linear Regression Model


```python
model = LinearRegression()
model.fit(X, y)
```

## Prediction


```python
predicted_score = model.predict([[6]])
predicted_score
```

## Visualization


```python
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.title("Linear Regression Example")
plt.show()
```

## Conclusion

This notebook demonstrates a basic linear regression workflow,
including data preparation, model training, prediction, and visualization.
