# Random Forest Regression Introduction

This notebook demonstrates a simple Random Forest regression workflow,
including model training and evaluation.


```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

## Create Sample Dataset


```python
data = {
    "StudyHours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Score": [50, 55, 65, 70, 80, 82, 88, 92]
}

df = pd.DataFrame(data)
df
```

## Train-Test Split


```python
X = df[["StudyHours"]]
y = df["Score"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

## Train Random Forest Model


```python
model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
model.fit(X_train, y_train)
```

## Model Evaluation


```python
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```

## Visualization


```python
plt.scatter(X_test, y_test, label="Actual")
plt.scatter(X_test, y_pred, label="Predicted")
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.title("Random Forest Regression Results")
plt.legend()
plt.show()
```

## Conclusion

This notebook demonstrates a basic Random Forest regression workflow,
including data splitting, model training, evaluation, and visualization.
